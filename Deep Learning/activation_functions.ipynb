```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
```

```python
# for output layer

def Sigmoid_b(x):
    return(1/(1+np.exp(-x)))

x = np.linspace(-5,5)
plt.plot(x, Sigmoid_b(x))
plt.title('activation -binary sig. fun')
plt.show()
```

```python
def TanH(x):
    return np.tanh(x)

x = np.linspace(-5,5)
plt.plot(x, TanH(x))
plt.title('activation -binary TanH fun')
plt.show()
```

```python
#for hidden layer (relu function)
def RectifiedLinearUnit(x):
    list=[]
    for i in x:
        if i>0:
            list.append(i)
        else:
            list.append(0)
    return list
x = np.linspace(-5,5)
plt.plot(x, RectifiedLinearUnit(x))
plt.title('activation  RectifiedLinearUnit fun')
plt.show()
```

```python
def LeakyreLu(x):
    list =[]
    for i in x:
        if i>0:
            list.append(i)
        else:
            list.append(0.01*i)
    return list
x = np.linspace(-5,5)
plt.plot(x, LeakyreLu(x))
plt.title('activation- LeakyreLu fun')
plt.show()
```

```python
def softmax(x):
    return (np.exp(x)/np.sum(np.exp(x)))


x = np.linspace(-5,5)
plt.plot(x, softmax(x))
plt.title('activation- softmax fun')
plt.show()
```
